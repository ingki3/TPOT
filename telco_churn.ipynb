```python
import pandas as pd
import numpy as np

telco_churn = pd.read_csv("data/Telco-Customer-Churn.csv")
telco_churn.head(5)
```

```python
telco_churn.shape
```

```python
telco_churn.dtypes
```

```python
column_list = telco_churn.columns

for column_name in column_list:
    if(telco_churn[column_name].dtype == 'object'):
        print("column name : {0}, unique items : {1}, items : {2} ".format(column_name, telco_churn[column_name].unique().size, telco_churn[column_name].unique()) )
```

```python
telco_churn['MultipleLines_YN'] = telco_churn['MultipleLines'].map({'No phone service' : 0, 'No' : 1, 'Yes' : 1})
telco_churn['InternetService_YN'] = telco_churn['InternetService'].map({'No' : 0, 'DSL' : 1, 'Fiber optic' : 1})
telco_churn['OnlineSecurity_YN'] = telco_churn['OnlineSecurity'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})
telco_churn['OnlineBackup_YN'] = telco_churn['OnlineBackup'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})
telco_churn['DeviceProtection_YN'] = telco_churn['DeviceProtection'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})
telco_churn['TechSupport_YN'] = telco_churn['TechSupport'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})
telco_churn['StreamingTV_YN'] = telco_churn['StreamingTV'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})
telco_churn['StreamingMovies_YN'] = telco_churn['StreamingMovies'].map({'No internet service' : 0, 'No' : 0, 'Yes' : 1})

telco_churn['gender'] = telco_churn['gender'].map({'Female' : 0, 'Male' : 1})
telco_churn['Partner'] = telco_churn['Partner'].map({'Yes' : 1, 'No' : 0})
telco_churn['Dependents'] = telco_churn['Dependents'].map({'Yes' : 1, 'No' : 0})
telco_churn['PhoneService'] = telco_churn['PhoneService'].map({'Yes' : 1, 'No' : 0})
telco_churn['PaperlessBilling'] = telco_churn['PaperlessBilling'].map({'Yes' : 1, 'No' : 0})

telco_churn.rename(columns={'Churn' : 'class'}, inplace=True)

telco_churn['class'] = telco_churn['class'].map({'Yes' : 1, 'No' : 0})



telco_churn['TotalCharges'] = telco_churn['TotalCharges'].replace(' ', np.nan)

telco_churn = telco_churn[telco_churn["TotalCharges"].notnull()]
telco_churn = telco_churn.reset_index()[telco_churn.columns]

# https://www.kaggle.com/pavanraj159/telecom-customer-churn-prediction
# Graph 그리는 나름대로의 library를 만들어 두는 것이 필요할 것으로 생각 됨.
# 개인이 만든 라이브러리 참조 방법 확인!!!

#telco_churn['TotalCharges'].isnull().any()
telco_churn['TotalCharges'] = [float(str(val)) for val in telco_churn['TotalCharges']]

telco_churn.head(5)
#telco_churn['TotalCharges']
```

```python
telco_churn['gender'].dtype
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

#gender_trans = mlb.fit_transform([{str(val)} for val in telco_churn['gender'].values])
multipleLines_trans = mlb.fit_transform([{str(val)} for val in telco_churn['MultipleLines'].values])
internetService_trans = mlb.fit_transform([{str(val)} for val in telco_churn['InternetService'].values])
onlineSecurity_trans = mlb.fit_transform([{str(val)} for val in telco_churn['OnlineSecurity'].values])
onlineBackup_trans = mlb.fit_transform([{str(val)} for val in telco_churn['OnlineBackup'].values])
deviceProtection_trans = mlb.fit_transform([{str(val)} for val in telco_churn['DeviceProtection'].values])
techSupport_trans = mlb.fit_transform([{str(val)} for val in telco_churn['TechSupport'].values])
streamTV_trans = mlb.fit_transform([{str(val)} for val in telco_churn['StreamingTV'].values])
streamingMovies_trans = mlb.fit_transform([{str(val)} for val in telco_churn['StreamingMovies'].values])
contract_trans = mlb.fit_transform([{str(val)} for val in telco_churn['Contract'].values])
paymentMethod = mlb.fit_transform([{str(val)} for val in telco_churn['PaymentMethod'].values])

telco_churn.head(5)
```

```python
multipleLines_trans
```

```python
telco_churn_new = telco_churn.drop(['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'customerID',
                                   'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod', 'class'],
                                   axis=1)
telco_churn_new.head(5)
```

```python
telco_churn_new = np.hstack((telco_churn_new.values, multipleLines_trans, internetService_trans, onlineSecurity_trans,
                            onlineBackup_trans, deviceProtection_trans, techSupport_trans, streamTV_trans, streamingMovies_trans,
                            contract_trans, paymentMethod))
telco_churn_new.shape
```

```python
telco_churn_class = telco_churn['class'].values
telco_churn_new[0]
```

```python
from sklearn.cross_validation import train_test_split

training_indices, validation_indices = training_indices, testing_indices = train_test_split(telco_churn.index, stratify = telco_churn_class, train_size=0.75, test_size=0.25)
training_indices.size, validation_indices.size
```

```python
from dask.distributed import Client

client = Client(n_workers=4, threads_per_worker=2)
client
```
